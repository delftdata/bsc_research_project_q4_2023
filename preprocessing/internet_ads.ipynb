```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("!!! Change this to your path to the Internet Advertisements Data Set (ad.data) !!!", low_memory=False)
df_column_names = pd.read_csv("labels/internet_ads.names")
```

```python
columns = df.columns
df.rename(
    columns=dict([(column, df_column_names["ad.names"][i]) for (i, column) in enumerate(columns)]),
    inplace=True)
df["ad"].replace("ad.", 1, inplace=True)
df["ad"].replace("nonad.", 0, inplace=True)
df.replace("   ?", np.nan, inplace=True)
```

```python
print("Shape before dropna:", df.shape)
df.dropna(inplace=True)
print("Shape after dropna:", df.shape)
```
