```python
import pandas as pd
```

```python
df = pd.read_csv("!!! Change this to your path to the Crop Mapping Using Fused Optical-Radar Data Set (WinnipegDataset.txt) !!!",
                  low_memory=False)
df_crop_names = pd.read_csv("labels/crop.names")
```

```python
for crop_code in range(1, df_crop_names.index.size + 1):
    df["label"].replace(crop_code, df_crop_names["crop_names"][crop_code - 1], inplace=True)
```
