```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("https://archive.ics.uci.edu/ml/machine-learning-databases/arrhythmia/arrhythmia.data", header=None,
                  low_memory=False)
df_column_names = pd.read_csv("labels/arrhythmia.names", header=None)
```

```python
df.rename(columns=dict([(i, df_column_names[0][i]) for i in range(df.shape[1])]), inplace=True)
df.replace("?", np.nan, inplace=True)
```

```python
print("Shape before dropna:", df.shape)
df.dropna(inplace=True)
print("Shape after dropna:", df.shape)
```
