```python
import pandas as pd
from sklearn.feature_selection import SequentialFeatureSelector
from sklearn.svm import LinearSVC
import random
```

```python
df = pd.read_csv("../../Datasets/Crop/WinnipegDataset.txt", low_memory=False)

for column in df.columns[1:]:
    if df[column].dtype != "float64":
        df[column] = df[column].astype("float")

df.head()
```

```python
X = df.iloc[random.sample(range(df.shape[0]), int(df.shape[0] * 0.001)), 1:]
y = df.iloc[random.sample(range(df.shape[0]), int(df.shape[0] * 0.001)), :1]
```

```python
import warnings
warnings.filterwarnings("ignore")

linear_svc = LinearSVC()
forward_selector = SequentialFeatureSelector(
    linear_svc, scoring="accuracy", n_features_to_select=2, direction="forward")
forward_selector.fit(X, y.values.ravel())

selected_features_mask = forward_selector.get_support()
selected_feature_indices = [i for (i, x) in enumerate(selected_features_mask) if x]

print("Before Forward feature selection:", df.shape)
print("After Forward feature selection:", df.iloc[:, selected_feature_indices + [-1]].shape)

df.iloc[:, [0] + selected_feature_indices].head()
```
